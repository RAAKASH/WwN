```python
import numpy as np
from graph import *
from timeit import default_timer as timer

def timecalc(func):
    def wrap(*args, **kwargs):
        t1 = timer()
        result = func(*args,**kwargs)
        t2 = timer()
        try:
            if kwargs["verbose"]=="on":
                print(result)
        except:
            print(result)
            

        return t2-t1
        
    return wrap  
```
